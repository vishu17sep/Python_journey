```python
import requests
import bs4
```

```python
res=requests.get("https://www.thegoldbugs.com/blog")
```

```python
soup=bs4.BeautifulSoup(res.text,"lxml")
```

```python
soup.select('pre')
```

```python
s=soup.select('pre')[0]
```

```python
type(s)
```

```python
tostr=s.getText()
l=tostr.split("-----")
l
```

```python
l.pop(0)
hint=""
for i in l:
    hint += i[0]
```

```python
hint
```

```python
res=requests.get(hint)
res.content
```

```python
soup=bs4.BeautifulSoup(res.content,'lxml')
soup
```

```python
file=open('hint.txt','w')
file.write(hint)
file.close()
```
